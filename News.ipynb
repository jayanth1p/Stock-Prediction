```python
ne
new=[]
for i in ne:
    d={}
    d.update({'title':i['title']})
    # d.update({"body":i['description'] or i['body']})
    new.append(d)
new
```

```python
sent_pol=[]
for i in news_ap:
    blob = TextBlob(i['title'])
    tse_pol = blob.sentiment.polarity
    blob = TextBlob(i['description'])
    dse_pol = blob.sentiment.polarity
    # print(tse_pol,dse_pol)
    sent_pol.append((tse_pol+dse_pol)/2)
sent_pol
```

```python
nlp = pipeline("sentiment-analysis")

# Step 4: Perform sentiment analysis on news articles
senti = []
for article in news_ap:
    sentiment_result = nlp(article['title'])
    sentiment = sentiment_result
    senti.append(sentiment)
senti
```

```python
sent_polar=[]
senti[0]
for i in senti:
    if i[0]['label']=="POSITIVE":
        sent_polar.append(i[0]["score"])
    else:
        sent_polar.append(-1*i[0]["score"])
```

```python
for i,j in zip(news_ap,sent_polar):
    print(i,j)
```
