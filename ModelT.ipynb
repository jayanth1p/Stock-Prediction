```python
from sklearn.preprocessing import MinMaxScaler
```

```python
import yfinance as yf
import pandas as pd
msft = yf.Ticker("PVP.NS")
```

```python
k=msft.history(period="3mo")
dataf=pd.DataFrame(k)
data=dataf.drop(columns=["Volume","Dividends","Stock Splits"])
data
```

```python
scaler = MinMaxScaler()
data_s = scaler.fit_transform(data)
data_s
```

```python
n_steps = 10  # You can adjust this based on your dataset and prediction horizon
n_features = 4

# Prepare the data for LSTM training
X, y = [], []
for i in range(len(data_s) - n_steps):
    X.append(data_s[i:i + n_steps])
    y.append(data_s[i + n_steps])
X
```

```python
pre=nmodel.predict( X)
pre
```
