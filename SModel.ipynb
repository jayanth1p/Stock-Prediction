```python
from datetime import datetime
```

```python
import yfinance as yf
import pandas as pd
msft = yf.Ticker("PVP.NS")
```

```python
k=msft.history(period="3y")
dataf=pd.DataFrame(k)
data=dataf.drop(columns=["Dividends","Stock Splits"])
data.corr()
```

```python
data
```

```python
dataf["2020-09-28":"2023-09-26"]
df=pd.DataFrame()
ls=[pd.date_range(start="2020-09-28",end="2023-09-26")]
lk=[]
for i in ls[0]:
    lk.append(i-ls[0][0])
len(lk)
```

```python
df=data
```

```python
df
df.reset_index(inplace=True)
df
```

```python
reference_date = datetime(2023, 1, 1)

# Function to calculate the number of days since the reference date
def days_since_reference_date(date):
    delta = date - reference_date
    return delta.days

# Apply the function to the date_column
df['days_since_reference'] = df['Date'].apply(days_since_reference_date)

# Display the DataFrame with the number of days column
print(df)
```
